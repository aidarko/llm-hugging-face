```python
from transformers import pipeline
import pandas as pd
```

```python
classifier = pipeline("text-classification")
# https://huggingface.co/SamLowe/roberta-base-go_emotions
# classifier = pipeline("text-classification", model = "SamLowe/roberta-base-go_emotions")

text = "i love you"
outputs = classifier(text)
pd.DataFrame(outputs)
```

# Named Entity Recognition

My name is Omar I live in Zurich -> Token Classification Model -> My name is Omar[Person] and I live in Zurich[City]

```python
ner_tagger = pipeline("ner", aggregation_strategy = "simple")
# finding importand keyphrase:
# ner_tagger = pipeline("ner", aggregation_strategy = "simple", model = "ml6team/keyphrase-extraction-kbir-inspec")
text = "My name is Omar. I live in Zurich."

output = ner_tagger(text)
pd.DataFrame(output)
```

# Question Answer Pipeline

text question -> question answering model -> text answer

```python
text = """
Dear Amazon, last week I ordered an Optimus Prime action figure from your online store in France. Unfortutely when I opened the package, I discovered that I had been sent an action figure of Saylor Moon.
"""
reader = pipeline("question-answering")
question = "what was wrong?"

output = reader (question = question, context = text)
pd.DataFrame([output])
```

# Summarization

```python
text = """
Dear Amazon, last week I ordered an Optimus Prime action figure from your online store in France. Unfortutely when I opened the package, I discovered that I had been sent an action figure of Saylor Moon.
"""
summarizer = pipeline("summarization")

output = summarizer(text)
pd.DataFrame(output)
```

# Text Generation

```python
generator = pipeline("text-generation") # gpt2-large

response = "I am sorry to hear that your order was mixed up"

output = generator(response, max_length=128)
print(output)
```
